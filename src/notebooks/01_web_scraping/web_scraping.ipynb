```python
import requests
import nest_asyncio
import asyncio
from bs4 import BeautifulSoup
from requests_html import AsyncHTMLSession
```

## Web Scraping Script

## cod vid spa

```python
session = AsyncHTMLSession()
nest_asyncio.apply()
```

```python
url = "https://react-amazon-bestsellers-books-dy.netlify.app/"
async def fetch_page():
    response = await session.get(url)
    await response.html.arender()
    return response
response = asyncio.run(fetch_page())
```

```python
print(response.content)
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
print(soup.find_all('article', class_="book"))
```

## cod do video

```python
pagina = requests.get('https://quotes.toscrape.com/')
dados_pagina = BeautifulSoup(pagina.text, 'html.parser')
```

```python
print(dados_pagina.prettify())
```

```python
todas_frases = dados_pagina.find_all("div", class_="quote")
```

```python
for div in todas_frases:
    texto = div.find("span", class_="text").text
    print(texto)
```

```python
todas_frases_prop = dados_pagina.find_all("span", itemprop="text")
```

```python
for div in todas_frases_prop:
    print(div.text)
```
